## 다중 선형회귀

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

X = np.array([[70,85,11], [71,89,18], [50,80,20],[99,20,10], [50,10,10]]) # 중간, 기말, 가산점
# 입력 벡터의 차원은 3 (input_dim = 3)
y = np.array([73, 82, 72, 57, 34]) # 최종 성적
# 출력 벡터의 차원은 1 (output_dim = 1)

model = Sequential()
model.add(Dense(1, input_dim = 3, activation = 'linear'))

sgd = optimizers.SGD(lr=0.00001)
model.compile(optimizer=sgd, loss='mse', metrics = ['mse'])
# sgd: 경사하강법
# loss function: mse(평균 제곱 오차)

model.fit(X,y, batch_size = 1, epochs = 2000, shuffle = False)
# 주어진 X와 y에 대해 오차를 최소화하는 작업을 2000번 시도
```

```python
print(model.predict(X))
print(y)
```

```python
X_test=np.array([[20,99,10], [40, 50, 20]]) # training에 사용하지 않은 데이터. 각각 58, 56점을 예측해야 함
print(model.predict(X_test))
```

## 다중 로지스틱 회귀

```python
import numpy as np
X = np.array([[0,0], [0,1], [1,0], [1,1]])
# 입력 벡터의 차원은 2. input_dim = 2
y= np.array([0,1,1,1])
# 출력 벡터의 차원은 1.output_dim = 1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
model.add(Dense(1, input_dim=2, activation = 'sigmoid')) # 입력의 차원은 2
model.compile(optimizer = 'sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])

model.fit(X,y,batch_size = 1, epochs = 800, shuffle = False)
```

```python
print(model.predict(X))
```
