## 2. Functional API로 만든 모델
각 층을 일종의 함수(function)로서 정의. 각 함수를 조합하기 위한 연산자 제공, 이를 이용해 신경망 설계.

### 1) Fully-connected FFNN
입력층 정의: 입력 데이터의 크기(shape)로 해야 함

```python
# 입력층 차원이 1인 Fully-connected FFNN (전결합 피드 포워드 신경망)이라 가정
from tensorflow.keras.layers import Input, Dense # 입력층 Input
from tensorflow.keras.models import Model
```

```python
inputs = Input(shape=(10,)) # 입력층 : 입력의 크기 정의
```

```python
hidden1 = Dense(64, activation = 'relu')(inputs) # 은닉층 : 이전 층을 다음 층 함수의 입력으로 사용하고 변수에 할당
hidden2 = Dense(64, activation = 'relu')(hidden1) # 은닉층 : 이전 층을 다음 층 함수의 입력으로 사용하고 변수에 할당
output = Dense(1, activation = 'sigmoid')(hidden2) # 출력층 : 이전 층을 다음 층 함수의 입력으로 사용하고 변수에 할당
```

```python
# 모델로 구성해보기
model = Model(inputs = inputs, outputs = output) # 이제 Sequential로 만든 모델처럼 model.compile, model.fit 등 사용 가능
```

### 2) Linear Regression

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import optimizers
from tensorflow.keras.models import Model

x = [1,2,3,4,5,6,7,8,9] # 공부 시간
y = [11,22,33,44,53,66,77,87,95] # 각 공부 시간에 맵핑되는 성적

inputs = Input(shape=(1,))
output = Dense(1, activation = 'linear')(inputs)
linear_model = Model(inputs, output)

sgd = optimizers.SGD(lr = 0.01)

linear_model.compile(optimizer = sgd, loss = 'mse', metrics = ['mse'])
linear_model.fit(x, y, batch_size = 1, epochs = 300, shuffle = False) # 참고로 verbose 디폴트가 1
```

```python
linear_model.predict([6]) # predict의 인자로 one single scalar가 들어가면 안됨
```

### 4) 다중 입력을 받는 모델

```python
model = Model(inputs = [a1, a2], outputs = [b1, b2, b3])
```

```python
# 다중 입력을 받는 모델을 설계해보자
from tensorflow.keras.layers import Input, Dense, concatenate
from tensorflow.keras.models import Model

# 두 개의 입력층 정의
inputA = Input(shape=(64,))
inputB = Input(shape=(128,))

# 첫번째 입력층으로부터 분기되어 진행되는 인공 신경망 정의
x = Dense(16, activation = 'relu')(inputA)
x = Dense(8, activation = 'relu')(x)
x = Model(inputs = inputA, outputs = x)

# 두번째 입력층으로부터 분기되어 진행되는 인공 신경망을 정의
y = Dense(64, activation = 'relu')(inputB)
y = Dense(32, activation = 'relu')(y)
y = Dense(8, activation = 'relu')(y)
y = Model(inputs = inputB, outputs = y)

# 두개의 인공신경망의 출력 연결 (concatenate!!!)
result = concatenate([x.output, y.output])

# 연결된 값을 입력으로 받는 밀집층 추가(Dense Layer)
z = Dense(2, activation = 'relu')(result)
# 선형 회귀를 위해 activation=linear 설정
z = Dense(1, activation = 'linear')(z)

# 결과적으로 이 모델은 두 개의 입력층으로부터 분기되어 진행된 후 마지막에 하나의 출력 예측하는 모델
model = Model(inputs = [x.input, y.input], outputs = z)
```
